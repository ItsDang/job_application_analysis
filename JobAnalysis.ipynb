```python
# imports
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import calendar
```

```python
# get the excel sheet
df = pd.read_excel('Jobs.xlsx', dtype={'Applied Date': datetime.datetime})
# print (df.head) 
```

```python
# Grab the days, convert to days of the week
appliedDates = df['Applied Date']
# https://stackoverflow.com/questions/9847213/how-do-i-get-the-day-of-week-given-a-date
appliedDatesDofW = pd.DataFrame(appliedDates.apply(lambda x: x.weekday()))
dfappliedDatesDofW = appliedDatesDofW.groupby(['Applied Date']).size()
# update index to days of the week, https://www.alpharithms.com/python-datetime-weekday-name-482616/
daynumber = [0,1,2,3,4,5,6]
weekday = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
dfappliedDatesDofW = dfappliedDatesDofW.rename(index=dict(zip(daynumber,weekday)))
dfappliedDatesDofW
dfappliedDatesDofW.plot(kind="bar")
# It's interesting to note that I haven't applied to any jobs on Saturday!
```

```python
# How many responded?
response = {'Responded': df['Status'].count(), 'Applied': df['Applied Date'].count()}
responsePd = pd.Series(data=response)
responsePd.plot.barh(stacked="true")
```
